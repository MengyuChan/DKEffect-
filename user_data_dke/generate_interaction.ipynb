```python
import os
import pandas as pd 
dog_data = pd.read_csv("/Users/emily/Desktop/categorization_experiment-master/data/dog_indiv_50_removed.csv") 
ball_data = pd.read_csv("/Users/emily/Desktop/categorization_experiment-master/data/bball_top50_decimal_removed.csv") 
```

```python
filename = '1616714995496-phase1_2021-03-25_230250_2021-03-26_000250.log'
file = open(filename, 'r')
lines = file.read().splitlines()
file.close()
```

```python
keep = []
interactions = ['help_hover', 'set_attribute_weight_vector_init', 'set_attribute_weight_vector_select', 'set_attribute_weight_vector_drag', 'set_attribute_weight_vector_calc', 'double_click', 'category_click', 'eventType\":\"click', 'eventType\":\"hover', 'eventType\":\"drag']
for line in lines:
    for interaction in interactions:
        if interaction in line:
            keep.append(line)
# keep[0:5]
```

```python
category = set()
for line in lines:
    if "category_click" in line:
        category.add(line.partition('\"dataItem\":\"')[2].partition('\",\"eventName')[0])
# category
```

```python
short = []
for kept in keep:
    string = kept.partition("whichCondition")[0] + kept.partition('eventSpecificInfo\":{},')[2]
    short.append(string)
# short[0:5]
```

```python
# test unique user IDs
uid = {}
for item in short:
    unique_id = (item.partition("userId\":\"")[2][0:13])
    if unique_id not in uid:
        uid[unique_id] = [item]
    else:
        uid[unique_id].append(item)
```

# correctness

```python
correctness = {}
uid_lst = []
for item in short:
    text = item.partition("classification\":\"")[2]
    classification = text.partition("\",\"userId")[0]
    userid = text.partition("\",\"userId\":\"")[2][0:13]
    name = item.partition("\"Name\":\"")[2].partition("\",\"")[0]
    if userid in uid.keys():
        if userid not in correctness.keys():
            correctness[userid] = [[name, classification]]
        else:
            correctness[userid].append([name, classification])

for user in correctness:
    for ans in correctness[user]:
        name = ans[0]
        classification = ans[1]
        if name[0] == "D":
            dog_id = name[4:]
            true_label = dog_data.loc[dog_data["id"] == int(dog_id)].iloc[0, 1]
            if classification == true_label:
                ans.append("T")
            elif classification == 'none':
                ans.append("NA")
            else:
                ans.append("F")
        if name[0] == "P":
            player_id = name[7:]
            true_label = ball_data.loc[ball_data["Player Anonymized"] == int(player_id)].iloc[0, 3]
            if classification == true_label:
                ans.append("T")
            elif classification == 'none':
                ans.append("NA")
            else:
                ans.append("F")
# correctness
```

# output csv

```python

for user_id in uid:
    for log in uid[user_id]:
        row = [user_id]
        # event type
        if 'help_hover' in log:
            row.append('help_hover')
        elif 'set_attribute_weight_vector_init' in log:
            row.append('set_attribute_weight_vector_init')
        elif 'set_attribute_weight_vector_select' in log:
            row.append('set_attribute_weight_vector_select')
        elif 'set_attribute_weight_vector_drag' in log:
            row.append('set_attribute_weight_vector_drag')
        elif 'set_attribute_weight_vector_calc' in log:
            row.append('set_attribute_weight_vector_calc')
        elif 'double_click' in log:
            row.append('double_click')
        elif 'category_click' in log:
            row.append('category_click')
        elif 'eventType\":\"click' in log:
            row.append('click')
        elif 'eventType\":\"hover' in log:
            row.append('hover')
        elif 'eventType\":\"drag' in log:
            row.append('drag')
        # elapsedTime
        row.append(log[log.find('elapsedTime\":')+len('elapsedTime\":'):log.rfind(',\"userId\"')]) if 'elapsedTime' in log else row.append('NA')
        # time stamp
        row.append(log.partition('eventTimeStamp\":\"')[2][0:-3])
        # taskType
        if 'set_attribute_weight_vector_drag' not in log:
            row.append(log[log.find('taskType\":\"')+len('taskType\":\"'):log.rfind('\",\"eventType\"')])
        elif "Dog" in log or "Apartment" in log or 'American' in log or 'Bernedoodle' in log or 'Charles' in log or 'Shih' in log:
            row.append("dog")
        elif '3-Pointers' in log or 'Player' in log or 'Center' in log or 'Guard' in log or 'Forward' in log:
            row.append('bball')
        else:
            row.append('NA')
        # now processing correctness - click
        if ("classification\":\"none" not in log and "eventType\":\"click" in log):
            name = log.partition("\"Name\":\"")[2].partition("\",\"")[0]
            classification = log.partition("classification\":\"")[2].partition("\",\"userId")[0]
#             print(name, classification)
#             row.append(name)
#             row.append(classification)
            if name[0] == "D":
                dog_id = name[4:]
                true_label = dog_data.loc[dog_data["id"] == int(dog_id)].iloc[0, 1]
#                 row.append(true_label)
                if classification == true_label:
                    row.append("T")
                else:
                    row.append("F")
            elif name[0] == "P":
                player_id = name[7:]
                true_label = ball_data.loc[ball_data["Player Anonymized"] == int(player_id)].iloc[0, 3]
                if classification == true_label:
                    row.append("T")
                else:
                    row.append("F")
        # now processing correctness - hover
        elif 'eventType\":\"hover' in log:
            name = log.partition("\"Name\":\"")[2].partition("\",\"")[0]
            if user_id in correctness.keys():
                for entry in correctness[user_id]:
                    if entry[0] == name:
                        row.append(entry[2])
                        break
            else:
                row.append("NA")
        # now processing correctness - drag
        elif 'eventType\":\"drag' in log:
            name = log.partition("\"Name\":\"")[2].partition("\",\"")[0]
            if user_id in correctness.keys():
                for entry in correctness[user_id]:
                    if entry[0] == name:
                        row.append(entry[2])
                        break
            else:
                row.append("NA")
        else:
            row.append("NA")
        output.append(row)
```
