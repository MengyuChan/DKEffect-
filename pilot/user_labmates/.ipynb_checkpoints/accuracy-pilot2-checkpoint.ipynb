```python
file = open('interaxis_2021-01-18_044545_2021-01-21_054545.log', 'r')
lines = file.read().splitlines()
file.close()
```

```python
keep = []
for line in lines:
    if "userlabel" in line and "Un-Assign" not in line and "hover" not in line and "drag" not in line:
        keep.append(line)
```

```python
short = []
for kept in keep:
    string = kept.partition("whichCondition")[0] 
    short.append(string)
```

```python
dict = {}
for item in short:
    text = item.partition("Name\":\"")[2]
    name = text.partition("\",\"coord")[0]
    rest = text.partition("classification\":\"")[2]
    classification = rest.partition("\",\"userId")[0]
    uid = rest.partition("\",\"userId")[2]
    uid = uid[3:-3]
    if uid in ["1611198166275", "1611100710666", "1611095788832", "1611099603264", "1611100826480"]:
        if uid not in dict:
            dict[uid] = [[name, classification]]
        else:
            dict[uid].append([name, classification])
```

```python
import csv 
```

```python
# loading data
import os
import pandas as pd 
dog_data = pd.read_csv("/Users/emily/Desktop/categorization_experiment-master/data/dog_indiv_100.csv") 
ball_data = pd.read_csv("/Users/emily/Desktop/categorization_experiment-master/data/bball_top100_decimal.csv") 
```

```python
# calculating accuracy
for user in dict:
    for ans in dict[user]:
        if ans[0][0] == "D":
            dog_id = ans[0][3:]
            user_label = ans[1]
            true_label = dog_data.loc[dog_data["id"] == int(dog_id)].iloc[0, 1]
            ans.append(true_label)
            if user_label == true_label:
                ans.append("T")
            else:
                ans.append("F")
        if ans[0][0] == "P":
            player_id = ans[0][7:]
            user_label = ans[1]
            true_label = ball_data.loc[ball_data["Player Anonymized"] == int(player_id)].iloc[0, 3]
            ans.append(true_label)
            if user_label == true_label:
                ans.append("T")
            else:
                ans.append("F")
```

```python
# accuracy
for user in dict:
    dog_count = 0
    dog_acc = 0
    ball_count, ball_acc = 0, 0
    for ans in dict[user]:
        if(ans[0][0]) == 'D':
            dog_count += 1
            if ans[-1] == 'T':
                dog_acc += 1
        if(ans[0][0]) == 'P':
            ball_count += 1
            if ans[-1] == 'T':
                ball_acc += 1
    print("user id:", user)
    print("dog:", dog_acc/dog_count)
    print("basketball:", ball_acc/ball_count) if ball_count != 0 else print("no basketball data")
    print()
```

```python
# saving the files for reference
for d in dict:
    name = d + '.csv'
    with open(name, 'w') as f: 
        write = csv.writer(f) 
        write.writerow(["name", "user_label", "true_label", "ans"])
        write.writerows(dict[d]) 
```
